```python
import OpenEXR
import numpy as np
import Imath

def read_exr(file_path,Channel):
    # Open the EXR file
    exr_file = OpenEXR.InputFile(file_path)

    # Get the header to obtain the data window
    header = exr_file.header()
    dw = header['dataWindow']
    size = (dw.max.x - dw.min.x + 1, dw.max.y - dw.min.y + 1)

    # Define the channels (assuming the depth map is in the 'Z' channel)
    FLOAT = Imath.PixelType(Imath.PixelType.FLOAT)
    channel = exr_file.channel(Channel, FLOAT)

    # Convert the data to numpy array
    depth_map = np.frombuffer(channel, dtype=np.float32)
    depth_map.shape = (size[1], size[0])  # (height, width)

    return depth_map
```

```python
import matplotlib.pyplot as plt

def display_depth_map(depth_map):
    plt.imshow(depth_map, cmap='viridis')
    plt.colorbar(label='Depth')
    plt.title('Depth Map')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()

```

```python
import numpy as np
import cv2
import OpenEXR
import Imath
from ultralytics import YOLO

# Function to read .exr depth file
def read_exr_depth(file_path):
    file = OpenEXR.InputFile(file_path)
    dw = file.header()['dataWindow']
    size = (dw.max.x - dw.min.x + 1, dw.max.y - dw.min.y + 1)
    pt = Imath.PixelType(Imath.PixelType.FLOAT)
    (R, G, B) = [np.frombuffer(file.channel(c, pt), dtype=np.float32) for c in "RGB"]
    R.shape = size
    G.shape = size
    B.shape = size
    return R, G, B

# Function to get depth values within the bounding box
def get_bbox_depth(R, bbox):
    xmin, ymin, xmax, ymax = map(int, bbox)
    return R[ymin:ymax, xmin:xmax]

# Function to calculate the centroid of a bounding box
def get_centroid(bbox):
    x_center = (bbox[0] + bbox[2]) / 2
    y_center = (bbox[1] + bbox[3]) / 2
    return int(x_center), int(y_center)

# Function to classify relative depth
def classify_relative_depth(relative_depth):
    if relative_depth < 0.25:
        return "L1 = VERY CLOSE"
    elif relative_depth < 0.5:
        return "L2 = CLOSE"
    elif relative_depth < 0.75:
        return "L3 = SAFE DISTANCE"
    else:
        return "L4 = FAR AWAY"



# Perform detection on the test image
image_path = '/content/drive/MyDrive/dataset/test/images/00035Right_png.rf.369a37d0aba1a04cbaab64838dfe7632.jpg'
depth_file_path = '/content/00035Right.exr'
results = model(image_path)

# Read the corresponding .exr depth file
R, G, B = read_exr_depth(depth_file_path)

# Load the RGB image for annotation
image = cv2.imread(image_path)

# Assuming only one result
result = results[0]
print(result)
if hasattr(result, 'boxes'):
    for box in result.boxes.xyxy:
        bbox = box[:4].tolist()  # [xmin, ymin, xmax, ymax]
        bbox_depths = get_bbox_depth(R, bbox)
        avg_depth = np.mean(bbox_depths)

        print(f'Bounding Box: {bbox}')
        print(f'Average Depth: {avg_depth}')

        # Calculate relative depth
        min_depth = np.min(R)
        max_depth = np.max(R)
        relative_depth = (avg_depth - min_depth) / (max_depth - min_depth)

        print(f'Relative Depth: {relative_depth}')

        # Classify relative depth
        depth_level = classify_relative_depth(relative_depth)
        print(f'Depth Level: {depth_level}')

        # Calculate centroid
        centroid = get_centroid(bbox)
        print(f'Centroid: {centroid}')

        # Annotate the image
        cv2.rectangle(image, (int(bbox[0]), int(bbox[1])), (int(bbox[2]), int(bbox[3])), (0, 255, 0), 2)
        cv2.circle(image, centroid, 5, (0, 0, 255), -1)
        cv2.putText(image, depth_level, (int(bbox[0]), int(bbox[1]) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

# Save the annotated image
cv2.imwrite('/content/drive/MyDrive/dataset/annotated_image1.jpg', image)

# Display the annotated image using cv2_imshow
from google.colab.patches import cv2_imshow
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
